```python
import pandas as pd
import numpy as np
import mne
import tensorflow as tf
import matplotlib.pyplot as plt
from mne.time_frequency import psd_array_multitaper
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras import layers, models
```

### Preparing the raw data

```python
def load_data(n):
    sub_raw_eeg = []
    sub_raw_data = []
    sub_raw_eeg_after = []
    sub_raw_data_after = []
    for i in range(n):
        for j in range(2):
            if i < 10:
                file = f"/content/drive/MyDrive/EEG data/eeg-during-mental-arithmetic-tasks-1.0.0/Subject0{i}_{j+1}.edf"
            else:
                file = f"/content/drive/MyDrive/EEG data/eeg-during-mental-arithmetic-tasks-1.0.0/Subject{i}_{j+1}.edf"

            data_after = mne.io.read_raw_edf(file, preload=True)  # Preload data to access it directly
            raw_data_after = data_after.get_data()

            if j == 0:
                sub_raw_data.append(data_after)
                sub_raw_eeg.append(raw_data_after)
            else:
                sub_raw_data_after.append(data_after)
                sub_raw_eeg_after.append(raw_data_after)

    return sub_raw_eeg, sub_raw_data, sub_raw_eeg_after, sub_raw_data_after

No_of_subjects = 16
array_list, data, array_after, data_after = load_data(No_of_subjects)

```

#  Power Spectral Density (PSD) Analysis

```python
def calculate_psd(data, sfreq, fmin, fmax):
    psds, freqs = psd_array_multitaper(data, sfreq=sfreq, fmin=fmin, fmax=fmax, verbose=0)
    return psds, freqs

bands = {'Delta': (1, 4), 'Theta': (4, 8), 'Alpha': (8, 12), 'Beta': (12, 30), 'Gamma': (30, 100)}

def band_psd_analysis(data, sfreq):
    band_psds = {band: [] for band in bands}
    for raw in data:
        for band, (fmin, fmax) in bands.items():
            psds, freqs = calculate_psd(raw, sfreq, fmin, fmax)
            band_psds[band].append(np.mean(psds, axis=1))
    return band_psds

sfreq = 170.5
band_psds_before = band_psd_analysis(array_list, sfreq)
band_psds_after = band_psd_analysis(array_after, sfreq)
```

Comparing PSD between two states

```python
def plot_band_psd(band_psds_before, band_psds_after, band_name):
    mean_psd_before = np.mean(band_psds_before[band_name], axis=0)
    mean_psd_after = np.mean(band_psds_after[band_name], axis=0)

    plt.figure(figsize=(10, 6))
    plt.plot(mean_psd_before, label='Rest State')
    plt.plot(mean_psd_after, label='Task State')
    plt.title(f'{band_name} Band PSD Comparison')
    plt.xlabel('Channels')
    plt.ylabel('PSD')
    plt.legend()
    plt.show()

# Compare and plot PSDs for each band
for band in bands.keys():
    plot_band_psd(band_psds_before, band_psds_after, band)
```

Step 3: Deep Learning Classification
 Data Preprocessing

```python
def preprocess_data(data, data_after, sfreq):
    # Resample data
    for raw in data:
        raw.resample(sfreq)
    for raw in data_after:
        raw.resample(sfreq)

    # Concatenate and label data
    frames = []
    for d_before, d_after in zip(data, data_after):
        df_before = d_before.to_data_frame().drop('time', axis=1)
        df_after = d_after.to_data_frame().drop('time', axis=1)
        df_before['label'] = 0
        df_after['label'] = 1
        frames.append(df_before)
        frames.append(df_after)

    model_input_data = pd.concat(frames)
    X = model_input_data.drop('label', axis=1).values
    y = model_input_data['label'].values
    return X, y

X, y = preprocess_data(data, data_after, sfreq)
```

 Spliting  data


```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.15, random_state=42)
```

Standardize data

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)
X_test = scaler.transform(X_test)
```

Reshape data for CNN models

```python
X_train = X_train.reshape(-1, X_train.shape[1], 1)
X_val = X_val.reshape(-1, X_val.shape[1], 1)
X_test = X_test.reshape(-1, X_test.shape[1], 1)
```

# Model 1: **EEGNet**

```python
def EEGNet_model(input_shape):
    model = models.Sequential()
    model.add(layers.Conv1D(16, 64, padding='same', input_shape=input_shape, activation='relu'))
    model.add(layers.BatchNormalization())
    model.add(layers.DepthwiseConv1D(1, depth_multiplier=2, activation='relu'))
    model.add(layers.BatchNormalization())
    model.add(layers.AveragePooling1D(2))
    model.add(layers.Dropout(0.5))
    model.add(layers.Conv1D(32, 16, padding='same', activation='relu'))
    model.add(layers.BatchNormalization())
    model.add(layers.AveragePooling1D(2))
    model.add(layers.Dropout(0.5))
    model.add(layers.Flatten())
    model.add(layers.Dense(1, activation='sigmoid'))

    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

# Model 2: **TSCeption**

```python
def TSCeption_model(input_shape):
    input_layer = layers.Input(shape=input_shape)
    conv1 = layers.Conv1D(filters=32, kernel_size=3, activation='relu')(input_layer)
    conv2 = layers.Conv1D(filters=64, kernel_size=3, activation='relu')(conv1)
    flat = layers.Flatten()(conv2)
    dense = layers.Dense(128, activation='relu')(flat)
    output_layer = layers.Dense(1, activation='sigmoid')(dense)

    model = models.Model(inputs=input_layer, outputs=output_layer)
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

```python
input_shape = (X_train.shape[1], 1)
```

Train EEGNet


```python
eegnet = EEGNet_model(input_shape)
history_eegnet = eegnet.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=50, batch_size=32)
```

Train TSCeption


```python
tsception = TSCeption_model(input_shape)
history_tsception = tsception.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=5, batch_size=32)
```

```python
# Evaluate Models
eegnet.evaluate(X_test, y_test)
tsception.evaluate(X_test, y_test)
```

```python
# Summarize Results
def plot_history(history, title):
    plt.plot(history.history['accuracy'], label='accuracy')
    plt.plot(history.history['val_accuracy'], label='val_accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.ylim([0, 1])
    plt.title(title)
    plt.legend(loc='lower right')
    plt.show()

plot_history(history_eegnet, 'EEGNet')
plot_history(history_tsception, 'TSCeption')
```
